```
def whitespace_tokenize(data):
  data = data.strip()    # 문자열의 맨앞, 맨끝 공백 지움
  if not data:
    return []
  tokens = data.split()  # 문자열을 스페이스,탭,엔터 단위로 분리하여 배열에 집어넣음
  return tokens
```

```
korquad_data = 'Downstream Task란? 딥러닝 관련 글들을 보면 최근 다운스트림 태스크에 대한 부분을 많이 볼 수 있다. 몇몇 글들을 찾아보다보면 잘 이해가지 않는 부분이 있었다가 한국어임베딩 책에서 적절한 설명을 찾았다. Downstream Task 의미 구체적으로 풀고 싶은 문제들을 말한다. 최근 자연어 처리분야에서는 언어모델을 프리트레인(pre-train)방식을 이용해 학습을 진행하고, 그 후에 원하고자 하는 태스크를 파인 튜닝(fine-tuning) 방식을 통해 모델을 업데이트 하는 방식을 사용하는데 이때, 태스크를 다운스트림 태스크라 한다. 예를들어, 버트의 언어모델을 질의응답 태스크인 squad를 학습한다고 할때, 이때 질의응답 태스크를 다운스트림 태스크로 볼 수 있을것이다. Refereces 도서-한국어임베딩'
# korquad_data = txt_file.read()
# txt_file.close()
```

```
from konlpy.tag import Mecab
tokenizer = Mecab()
    
output_tokens = [ '[PAD]', '[UNK]', '[CLS]', '[SEP]', '[MASK]' ]
# wst : 공백,탭,엔터 기준 문자열 하나
for wst in whitespace_tokenize(korquad_data):
  count = 0
  for token in tokenizer.morphs(wst):
    tk = token
    
    if count > 0:
      tk = "##" + tk
      if tk in output_tokens:
        continue
      output_tokens.append(tk)
    else:
      count += 1
      if tk in output_tokens:
        continue
      output_tokens.append(tk)  # 맨 처음 token만 앞에 ##을 붙이지 않음 
```
